```python
import os
import pandas as pd
import codecs
from pydub import AudioSegment
import refine_text
```

```python
vinbrain_folder = os.path.join('..', 'data', 'vinbrain')
```

```python
f = []
for (dirpath, dirnames, filenames) in os.walk(vinbrain_folder):
    f.extend(filenames)
    break
```

```python
audio = []
gcp = []
vais = []
gold = []
name = []

for i in f:
    ext = i.split('.')[-1]
    f_name = i.split('.')[0]
    if ext == 'wav':
        audio.append(i)
        name.append(f_name)
```

```python
r = []
for k, i in enumerate(name):
    ff_loc = os.path.join(vinbrain_folder, i + '_google.txt')
    if os.path.isfile(ff_loc):
        gc_txt = codecs.open(ff_loc, 'r', 'UTF-8').read()
    else:
        gc_txt = None
    
    ff_loc = os.path.join(vinbrain_folder, i + '_vais.txt')
    if os.path.isfile(ff_loc):
        va_txt = codecs.open(ff_loc, 'r', 'UTF-8').read()
    else:
        va_txt = None
    
    ff_loc = os.path.join(vinbrain_folder, i + '_gold.txt')
    if os.path.isfile(ff_loc):
        go_txt = codecs.open(ff_loc, 'r', 'UTF-8').read()
    else:
        go_txt = None
        
    f_audio = AudioSegment.from_file(os.path.join(vinbrain_folder, audio[k]))
        
    r.append((i, f_audio.duration_seconds, go_txt, va_txt, gc_txt))
```

```python
result = pd.DataFrame(r, columns=['id', 'duration', 'gold',  'vais', 'google'])
```
